# Retina Parameterisation

```python
num_nodes = 4096 * 2
fovea = 0.1
```

# Standard SSNN Retina

```python
retina = SSNN(num_nodes, fovea)
```

```python
retina.fit()
```

```python
retina.display_tessellation()
```

```python
retina.display_stats()
```

# Hybrid SSNN-Fibonacci Retina

```python
hybrid_retina = fib_ssnn_hybrid(num_nodes, fovea*0.7, foveal_density=1.6)
```

```python
display_tessellation(hybrid_retina)
```

```python
display_stats(hybrid_retina)
```

# Generative Retina

```python
gen_retina = SSNN(num_nodes, fovea)
```

```python
gen_retina.generative_fit(steps=1)
```

```python
gen_retina.display_tessellation()
```

```python
gen_retina.display_stats()
```

# Fibonacci Retina

```python
fib_retina = fibonacci_retina(num_nodes, fovea, fovea_density=2)
```

```python
display_tessellation(fib_retina)
```
